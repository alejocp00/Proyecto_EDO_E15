Metodo de Euler.

Sea la funcion:

```python
def funcion(x,y):
    return x + 1/5 * y
```

El metodo de Euler consiste en aplicar la siguiente formula:
y1 = y0 + h * f(x0,y0)
donde (x0,y0) son las coordenadas del punto anterior y (x1,y1) son las coordenadas del punto resultante. La h es el tamaño de paso y define tambien el incremento en las x de forma que:
x(n + 1) = x(n) + h
Cabe destacar que este metodo da una aproximación de la función en un intervalo y por tanto se comete un error al usarlo.
El siguiente código representa el método de Euler en python:

```python
import matplotlib.pyplot as math
def MetodoEuler(abs, y, func, h):
    ord = list() #lista con las ordenadas de la funcion
    ord.append(y) #agrego la ordenada que me dan
    
    MetodoEuler2(abs[0], y, func, h, abs[0], abs[len(abs) - 1], ord) #aplico la formula del metodo de Euler
    return ord #retorno la lista de las ordenadas para poder plotear la funcion

def MetodoEuler2(x, y, func, h, menor, mayor, ord):
    if x < menor or x >= mayor: return y #restrinjo la x para el intervalo que me dan
    b = y + h*func(x,y) #aplico la formula
    ord.append(b) #agrego el resultado a la lista
    return MetodoEuler2(x + h, b, func, h, menor, mayor, ord) 
```
